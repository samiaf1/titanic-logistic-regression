```python
df = pd.read_csv("train_and_test2 2.csv")
df.head()
```

```python
df.drop(columns=['zero','zero.1','zero.2','zero.3','zero.4','zero.5','zero.6','zero.7','zero.8','zero.9','zero.10','zero.11','zero.12','zero.13','zero.14','zero.15','zero.16','zero.17','zero.18'],inplace=True)
df.head()
```

```python
print(df.isnull().sum())
```

```python
df['Embarked'].fillna(df['Embarked'].mean(),inplace = True)
```

```python
print(df.isnull().sum())
```

```python
df.head()
```

```python
x = df.drop('2urvived',axis = 1)
y = df['2urvived']
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
print(x_train.shape,x_test.shape)
```

```python
moodel.fit(x_train,y_train)
```

```python
y_pred = moodel.predict(x_test)
```

```python
acc = accuracy_score(y_pred,y_test)
print("accuracy:",acc)
```

```python
y_prob = moodel.predict_proba(x_test)
loss = log_loss(y_test, y_prob)
print("log loss",loss)
```
